```python
x = df.drop('Price', axis=1)
x.head()
```

```python
y = df.drop('Price', axis=1)
y.head()
```

Linera Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(x,y)
```

```python
reg.coef_ # m
```

```python
reg.intercept_ # c
```

```python
reg.predict([[6]])
```

```python
df['residuals'] = df[['Price']] - reg.predict(x)
df.head()
```

```python
df['predict'] = reg.predict(x)
df.head(8)
```

```python
plt.scatter(df.Weight , df.predict, marker='o', color='black')
plt.xlabel('Weight')
plt.ylabel('Price')
```

```python
plt.scatter(df.Weight , df.Price, marker='o', color='red')
plt.scatter(df.Weight , df.predict, marker='o', color='black')
plt.xlabel('Weight')
plt.ylabel('Price')
```

```python
plt.scatter(df.Weight , df.Price, marker='o', color='red')
plt.plot(df.Weight, df.predict)
plt.xlabel('Weight')
plt.ylabel('Price')
```

MSE

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
mse = mean_squared_error(df['Price'] , df['predict'])
mse
```

```python
mean_absolute_error(df['Price'] , df['predict'])
```

```python
import math
math.sqrt(mse) #RMSE
```
